```python
import numpy as np
import matplotlib.pyplot as plt

# Function 1: Gaussian with mean at x = 1
x = np.linspace(-5, 5, 1000)
mean = 1  # Mean value
std_dev = 0.5  # Standard deviation
y_gaussian = np.exp(-(x - mean) ** 2 / (2 * std_dev ** 2))

# Scaling the Gaussian to range from 1 to 2 in the y-direction
y_gaussian = y_gaussian * (2 - 1) + 1

# Function 2: Line with the inclination of the derivative of sin(x)/2 + 1/2 at x = 1
slope = np.cos(1) / 2  # Slope of the derivative at x = 1
intercept = -slope * 1 + (np.sin(1) / 2 + 1 / 2)  # Calculate the y-intercept
y_line = slope * x + intercept

# Function 3: sin(x)/2 + 1/2
y_sin = np.sin(x) / 2 + 1 / 2

# Plotting the functions
plt.plot(x, y_gaussian, label='Gaussian (mean at x=1)')
plt.plot(x, y_line, label='Line with derivative inclination (crossing at x=1)')
plt.plot(x, y_sin, label='sin(x)/2 + 1/2')

# Plotting the evaluated point
plt.scatter(1, np.sin(1) / 2 + 1 / 2, color='red', label='Evaluated Point (x=1)')

# Set axis limits and labels
plt.xlim(0, 5)
plt.ylim(0, 2)
plt.xlabel('x')
plt.ylabel('y')

# Add a legend
plt.legend()

# Display the plot
plt.show()
```
