```python
import numpy as np
import matplotlib.pyplot as plt

class MyClass:
    def __init__(self):
        self.INTERVAL = 10
        self.X_train = None

    def xy_discrete(self, xy):
        return np.round(xy)

    def generate_training_data(self):
        self.X_train = self.xy_discrete(np.random.uniform(0, self.INTERVAL, size=(2, 2)))

    def visualize_data_structure(self):
        if self.X_train is None:
            print("No data available. Generate training data first.")
            return

        fig = plt.figure()
        ax = fig.add_subplot(111, projection='3d')

        x = self.X_train[:, 0]
        y = self.X_train[:, 1]
        z = np.zeros_like(x)

        ax.scatter(x, y, z, c='b', marker='o')

        ax.set_xlabel('X')
        ax.set_ylabel('Y')
        ax.set_zlabel('Z')

        plt.show()

# Create an instance of MyClass
my_object = MyClass()

# Generate training data
my_object.generate_training_data()

# Visualize the data structure
my_object.visualize_data_structure()
```
